```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import nltk
# from nltk.tokenize import word_tokenize
import spacy
from spacy import load
from spacy.lang.ru.examples import sentences
from spacy.lang.ru import Russian
from nltk.corpus import stopwords
```

```python
# nltk.download('punkt')
```

```python
file = open("bts.txt")
text = ' '.join(file.read().splitlines())
text
```

```python
text = re.sub("[A-Za-z0-9!#$%&()*+,./:;«»<=>?@[\]^_`{|}~—–\\-]+", '', text)
text = text.split()
```

Произведём лемматизацию

```python
nlp = Russian()
load_model = load("ru_core_news_sm")
nltk.download('stopwords')
stopwords_ru = stopwords.words("russian")
```

```python
stopwords_ru.append('е')
stopwords_ru.append('т')
stopwords_ru.append('х')
stopwords_ru.append('ба')
stopwords_ru.append('±')
stopwords_ru.append('°')
stopwords_ru.append('х')
stopwords_ru.append('×')
stopwords_ru.append('θ')
```

```python
lemma = []

for doc in load_model.pipe(text):
  for word in doc:
    lemma.append(word.lemma_)
```

```python
text = ' '.join(lemma)
```

Удаляем стоп-слова

```python
tokens = []
for token in text.split():
  if token not in stopwords_ru:
    tokens.append(token)
```

```python
words = list(set(tokens))
len(words)
```

```python
df = pd.DataFrame(0, index=words, columns=words)
df
```

```python
range_ = [2, 1, -1, -2]
for i, word in enumerate(tokens):
  if i == 0:
    df.loc[word, tokens[i + 1]] += 1
    df.loc[word, tokens[i + 2]] += 1
  elif i == 1:
    df.loc[word, tokens[i - 1]] += 1
    df.loc[word, tokens[i + 1]] += 1
    df.loc[word, tokens[i + 2]] += 1
  elif i == len(tokens) - 2:
    df.loc[word, tokens[i - 2]] += 1
    df.loc[word, tokens[i - 1]] += 1
    df.loc[word, tokens[i + 1]] += 1
  elif i == len(tokens) - 1:
    df.loc[word, tokens[i - 2]] += 1
    df.loc[word, tokens[i - 1]] += 1
  else:
    for j in range_:
      df.loc[word, tokens[i - j]] += 1

for i in range(len(tokens)):
  if df.loc[tokens[i], tokens[i]] > 0:
    df.loc[tokens[i], tokens[i]] = 0
```

```python
A_T = np.transpose(A)
```
